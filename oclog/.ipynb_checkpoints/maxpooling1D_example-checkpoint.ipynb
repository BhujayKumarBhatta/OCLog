```python
import tensorflow as tf
```

```python
x = tf.constant([1., 2., 3., 4., 5.])
x
```

```python
x = tf.reshape(x, [1, 5, 1])
x
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=2,
   strides=1, padding='valid')
max_pool_1d(x)
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=5,
   strides=1, padding='valid')
max_pool_1d(x)
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=2,
   strides=2, padding='valid')
max_pool_1d(x)
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=2,
   strides=1, padding='same')
max_pool_1d(x)
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=2, padding='same', strides=2)
max_pool_1d(x)
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=5, padding='same', strides=2)
max_pool_1d(x)
```

```python
input_shape = (3, 2, 4, 3)
x = tf.random.normal(input_shape)
x
```

```python
chars_in_line = 256
lines_in_seq = 64
char_onehot = 51
input_shape = (lines_in_seq, chars_in_line, char_onehot)
x = tf.random.normal(input_shape)
print(x.shape)
print(input_shape[2:])
```

```python
y = tf.keras.layers.Conv1D(
32, 3, activation='relu', input_shape=(64, 256, 51))(x)
y.shape
```

```python
max_pool_1d = tf.keras.layers.MaxPooling1D(pool_size=2,
   strides=1, padding='same')
max_pool_1d(y)
```

```python
#example of how to apply max pooling with 
input_shape = (4, 7, 10, 128)
print(input_shape[2:])
x = tf.random.normal(input_shape)
print(x.shape)
y = tf.keras.layers.Conv1D(
    32, 3, activation='relu', input_shape=input_shape[2:])(x)
print(y.shape)
y = tf.keras.layers.MaxPooling2D(pool_size=(1,8))(y)
y.shape
```
